## Learn about numpy 

```python
import numpy as np
```

 ### Creating array from list 

```python
np_array1 = np.array([1,2,3])
print("The array 1 is :",np_array1)

np_array2 = np.array([[1,2,3],[5,6,7]])
print("The array 2 is :",np_array2)
```

### Array vs List


```python
py_list = [1,2,3]
print("Python list multiplication :", py_list *2)

np_array = np.array([1,2,3])  # element wise multiplication
print("Python array multiplication :", np_array *2)

import time 

start_time = time.time()
py_list = [i*2 for i in range(1000000)]
print("\n List operation time : ",time.time()-start_time)

start_time = time.time()
np_array = np.arange(100000) * 2
print("\n Array operation time : ",time.time()-start_time)

```

### Creating array from scratch 

```python
zero = np.zeros((3,4))
print("Zeros array : \n",zero)

ones = np.ones((2,3))
print("ones array : \n",ones)

full = np.full((2,2),7)    # ((para1,para2),number to print in array)
print("Full array : \n",full)

random = np.random.random((4,3))  # ((para1,para2))
print("Random array : \n",random)

sequence = np.arange(0,10,2)    #(start,stop,step)
print("Sequence array : ",sequence)

```

### Array Properties 

```python
arr = np.array([[1,2,3],[4,5,6]])

print ("Shape :" ,arr.shape)
print ("Dimension : ",arr.ndim)
print ("Size : ",arr.size)
print ("Datatype : ",arr.dtype)
```

### Array Reshaping 

```python
arr = np.arange(12)
print("Original array :",arr)

reshaped = arr.reshape((3,4))
print("Reshaped array : \n ",reshaped)

flattened = reshaped.flatten()
print("Flatened array : ",flattened)

# raveled returns the view(original) not the copy 
raveled = reshaped.ravel()
print("Raveled array : ",raveled)

#Transpose 
transpose = reshaped.T
print("Tranpose array : \n", transpose)






```
